Therefore, based on a set of purchases made by the roommates for the house, each roommate will either end up with a credit, a debit, or a €0 final balance.

This function will take two parameters:

Otherwise (if the input does not satisfie the requirements), an error will be printed and an **empty list** will be returned.

````python
#// BEGIN_TODO [get_individual_totals] Get individual totals 
from typing import List

def get_individual_totals(number_of_people: int, purchases: list[any])-> list[any]:
    """
    Calculates total spent by each individual.
    
    Asumptions: The first arguments is the number of households and this value is a int.
                The second one is the list of purchases and this value is an list of any.
                
                
    :number_of_people: Is the number off people in a household
    :purchases: is a list of purchases by each individual
    
    Returns: Returns a list of total spent by each individual
    
    Example and test cases: 
    
    >>> get_individual_totals(5, [5.0, 1, 1.2, 2.4, 6.77, 0, 3.0, 2]) #invalid input
    Invalid input: Wrong concatenation of costs and ids in parameter 'purchases'.
    
    >>> get_individual_totals(2, [15.00, 0, 23.15, 0]) #valid input
    [38.15, 0.0]
    
    >>> get_individual_totals(5, [3.0, 1, 2.2, 1.4, 6.87, 0, 2.0, 2]) #invalid input
    Invalid input: Wrong concatenation of costs and ids in parameter 'purchases'.
    
    >>> get_individual_totals(2, [19.00, 0, 33.25, 0]) #valid input
    [52.25, 0.0]
    
    
    """
    
    final_balances = [0.0] * number_of_people

    # Check if the length of the purchases list is even (cost followed by ID).
    if len(purchases) % 2 != 0:
        print("Invalid input: Number of elements in 'purchases' should be even.")
        return []

    # Iterate through the purchases list by pairs (cost, ID).
    for i in range(0, len(purchases), 2):
        cost = purchases[i]
        person_id = purchases[i + 1]

        # Check if the person_id is within the valid range.
        if not (0 <= person_id < number_of_people):
            print("Invalid input: Person ID out of range.")
            return []

        # Update the final balance for the person.
        final_balances[person_id] += cost

    return final_balances

#// END_TODO [get_individual_totals]

get_individual_totals(2, [15.00, 0, 23.15, 0])
````

## Task 2: Final Balances

We then create a second function that 
- takes the total spent by each individual: a list of floats, where the float number at index `i` corresponds to the total spent by the person with ID `i`
- returns the final balance of each individual: a list of floats, where the float number at index `i` corresponds to the final balance of the person with ID `i`

**Notes**:
* The final balance of an individual can be calculated by doing `spent - avg`, where
    * `spent` is the total amount spent by the individual
    * `avg` is the total amount spent by the group, divided by the number of people in the group
* Remember to deal with the possibily that the input list is empty
* Do not forget to write the **type hints** and **docstring** of the function.
* Include at least _two_ (2) `doctest` examples

**Examples**:

*Input:*  
```python
get_individual_final_balances([10.0, 20.0, 3.0])
```

*Output:*  

```python
[-1.0, 9.0, -8.0]
```

````python
#// BEGIN_TODO [get_individual_final_balances] Get individual final balances


def get_individual_final_balances(total_spent_by_individual: list[float]) -> list[float]:
    """
    Calculate the final balance of each individual based on their total spending.

    Args:
    total_spent_by_individual (List[float]): A list of total amounts spent by each individual.

    Returns:
    List[float]: A list of final balances for each individual, calculated as total spent - avg,
    where total spent is the total amount spent by the individual and avg is the average spending
    by the group.

    If the input list is empty, an empty list is returned.

    Examples:
    >>> get_individual_final_balances([10.0, 20.0, 3.0])
    [3.333333333333333, -6.666666666666667, -10.0]

    >>> get_individual_final_balances([])
    []
    """
    # Check if the input list is empty.
    if not total_spent_by_individual:
        return []

    # Calculate the average spending by the group.
    avg = sum(total_spent_by_individual) / len(total_spent_by_individual)

    # Calculate the final balance for each individual.
    final_balances = [spent - avg for spent in total_spent_by_individual]

    return final_balances





#// END_TODO [get_individual_final_balances]

get_individual_final_balances([12.0, 10.0, 14.0])
````

````python
#// BEGIN_TODO [simplified_splitser] Create wrapper function for Simplified Splitser



def simplified_splitser(num_people: int, purchases: list[float, int]) -> list[float]:
    """
    Calculate the final balances of each individual in a household.

    Args:
    num_people (int): The number of people in the household.
    purchases (List[Union[float, int]]): A list of purchases, where each expense (float) is followed by the ID
            of the person who made the purchase (integer), which needs to be in the right range (greater or equal
            to 0 and less than the number of people in the household).

    Returns:
    List[float]: A list of final balances for each person in the household.

    If the input does not satisfy the requirements, errors will be printed from the 'get_individual_totals' function,
    and an empty list will be returned.

    Examples:
    >>> simplified_splitser(5, [12.45, 1, 15.6, 4, 25.5, 2, 4.30, 1])
    [-11.57, 5.18, 13.93, -11.57, 4.03]

    >>> simplified_splitser(2, [5.0, 1, 1.2, 2.4, 6.77, 0, 3.0, 2])
    Invalid input: Person ID out of range.
    []
    """
    try:
        # Calculate individual totals using get_individual_totals.
        individual_totals = get_individual_totals(num_people, purchases)

        # Calculate final balances using get_individual_final_balances.
        final_balances = get_individual_final_balances(individual_totals)

        return final_balances
    except ValueError as e:
        print(e)
        return []




#// END_TODO [simplified_splitser]

simplified_splitser(5, [12.45, 1, 15.6, 4, 25.5, 2, 4.30, 1])
````

### 3.2: Data Conversion
We now create another function that instead of two parameters only takes 1 parameter:

This means that the number of people and their purchases will be merged into one list of strings and it is our job to figure out how to
- transform the list into parameters that are fit for `simplified_splitser`
- **print** the output of `simplified_splitser` as multiple lines with 
    - each line having format "[name]: € [cost]" and 
    - the cost having **2 decimal points** (.2f).

**Notes:** 

- To check if a string `my_string` can be converted to an int, you can use `my_string.isdigit()`, this outputs a boolean. 
- To check if a string `my_string` can be converted to a float, you can use `my_string.replace('.', '', 1).isdigit()`, this outputs a boolean.

````python
#// BEGIN_TODO [string_list_simplified_splitser] Convert single string list into parameters for simplified_splitser



def string_list_simplified_splitser(input_strings: List[str]) -> None:
    """
    Transform a list of strings into parameters for simplified_splitser and print the results.

    Args:
        input_strings (List[str]): A list of strings containing information about the number of people and purchases.

    This function converts the input strings into parameters for simplified_splitser and prints the final balances
    for each person in the household.

    If the input strings cannot be correctly converted to parameters or if there are errors in simplified_splitser,
    those errors will be printed.

    Examples:
        >>> string_list_simplified_splitser(
        ...     ["5", "12.45", "1", "15.6", "4", "25.5", "2", "4.30", "1"]
        ... )
        0: € -11.57
        1: € 5.18
        2: € 13.93
        3: € -11.57
        4: € 4.03

        >>> string_list_simplified_splitser(
        ...     ["2", "5.0", "1", "1.2", "2.4", "6.77", "0", "3.0", "2"]
        ... )
        0: € 3.00
        1: € -3.00

        >>> string_list_simplified_splitser(["InvalidInput", "12.45", "1", "15.6", "4", "25.5", "2", "4.30", "1"])
        Invalid input: Wrong concatenation of costs and ids in parameter 'purchases'
        invalid literal for int() with base 10: 'InvalidInput'
    """
    try:
        # Extract the number of people from the first string in the list.
        num_people = int(input_strings[0])

        # Extract the purchases as a list of floats and integers.
        purchases = []
        for i in range(1, len(input_strings)):
            if i % 2 == 1 and input_strings[i].replace('.', '', 1).isdigit():
                purchases.append(float(input_strings[i]))
            elif i % 2 == 0 and input_strings[i].isdigit():
                purchases.append(int(input_strings[i]))
            else:
                raise ValueError("Invalid input: Incorrect format in input strings.")

        # Call simplified_splitser to get the final balances.
        final_balances = simplified_splitser(num_people, purchases)

        # Print the final balances with formatting.
        for i, balance in enumerate(final_balances):
            print(f"{i}: € {balance:.2f}")
    except ValueError as e:
        print("Invalid input: Wrong concatenation of costs and ids in parameter 'purchases'")
        print(e)




#// END_TODO [string_list_simplified_splitser]


string_list_simplified_splitser(
    ["5", "12.45", "1", "15.6", "4", "25.5", "2", "4.30", "1"]
)
````

## Task 4: Process Text File

In order to do this, we now write a function `process_text_file` that 
- Opens the text file whose file name is given by `FILE_NAME`,

````python
FILE_NAME = 'assets/simplified_splitser.txt'

def process_text_file(file_name: str) -> None:
#// BEGIN_TODO [process_file] Process household expenses in text file 
 """
    Process a text file containing household expense data.

    Arguments:FILE_NAME (str): The name of the text file to be processed.

    This function opens the specified text file, reads it line by line, and processes each household's data using
    string_list_simplified_splitser. It prints the results with one blank line in between.

    Example:
    >>> process_text_file("assets/simplified_splitser.txt")
        0: € 0.00

        0: € 0.14
        1: € -0.14
    """
try:
    with open(FILE_NAME, "r") as file:
        lines = file.readlines()

        # Process each line in the file.
    for line in lines:
        # Strip leading and trailing whitespace and split into a list of strings.
        data = line.strip().split()

        # Call string_list_simplified_splitser to process the household's data.
        string_list_simplified_splitser(data)

        # Print a blank line between households.
        print()
except FileNotFoundError:
        print(f"File '{FILE_NAME}' not found.")
except Exception as e:
        print(f"An error occurred: {e}")
    


#// END_TODO [process_file] 

process_text_file(FILE_NAME)
````

````python
#// BEGIN_TODO [process_input_value] Process household expenses input data ()

def process_input_value() -> None:
    """
    Repeatedly read and process user input for household expense data.

    This function continuously prompts the user for input, processes the data using string_list_simplified_splitser,
    and prints the final balances. The program is terminated if the user input is 'q' or 'quit'.

    Example:
        >>> process_input_value()
        Enter household data (or 'q' to quit): 3 2.3 2 10.5 0
        0: € 6.23
        1: € -4.27
        2: € -1.97
        Enter household data (or 'q' to quit): 2 230.7 0 100.0 1 74.5 1 0.90 0 58.9 1
        0: € -0.90
        1: € 0.90
        Enter household data (or 'q' to quit): 3 24.5 0 20.3 2 3.5 2 41.2 1 3.4
        Invalid input: The list of purchases does not have even length.
        Enter household data (or 'q' to quit): 2 4.5 2 10.2 1 3.7 0
        Invalid input: The ID at index 1 in purchases is not in range.
        Enter household data (or 'q' to quit): q
    """
    while True:
        user_input = input("Enter household data (or 'q' to quit): ").strip()

        if user_input.lower() in {'q', 'quit'}:
            break

        if not user_input:
            print("No data inputted.")
            continue

        data = user_input.split()
        result = string_list_simplified_splitser(data)

        if result:
            for i, balance in enumerate(result):
                print(f"{i}: € {balance:.2f}")
        else:
            print("Invalid input. Please try again.")

#// END_TODO [process_input_value]
        
prompt = "Insert household names and expenses: "
while True:
    inp = input(prompt)
    if(inp.lower() == "quit" or inp.lower() == "q"):
        break
    process_input_value(inp)
````

## Task 6: Drawing a Bar Chart (Bonus)

To visualize the final balances, we create a function to show them in a bar chart. The x-axis of the plot will show the ids of the group members. The y-axis of the plot will contain the corresponding balances.

Luckily, it is quite easy to visualize (or draw) the data from the output of the function `simplified_splitser` as a bar chart.

The negative balances should be colored in orange, while the positive balances in blue
 - this can be achieved by creating a list `colors` where each color is based on whether the value of a key is positive or negative
   

**Examples**:

*Input:*

```python
data = simplified_splitser(5, [12.45, 1, 15.6, 4, 25.5, 2, 4.30, 1])
bar_chart(data)
```

*Output:*

````python
#// BEGIN_TODO [bar_chart] Drawing a bar chart


def bar_chart(final_balances: list[float]) -> None:
    """
    Create and display a bar chart to visualize final balances.

    Arguments:final_balances (List[float]): A list of final balances for each individual.

    This function creates a bar chart to visualize the final balances, with positive balances in blue and
    negative balances in orange. The x-axis shows the IDs of group members, and the y-axis contains the balances.

    Example:
>>> data = simplified_splitser(5, [12.45, 1, 15.6, 4, 25.5, 2, 4.30, 1])
>>> bar_chart(data)
        (This will display the bar chart)
    """
    # Determine the colors for the bars (blue for positive, orange for negative).
    colors = ['blue' if balance >= 0 else 'orange' for balance in final_balances]

    # Create a bar chart with IDs on the x-axis and balances on the y-axis.
    plt.bar(range(len(final_balances)), final_balances, color=colors)
    plt.xlabel("Member ID")
    plt.ylabel("Balance")
    plt.title("Final Balances Bar Chart")

    # Show the bar chart.
    plt.show()

# Test case
if __name__ == "__main__":
    data = simplified_splitser(5, [12.45, 1, 15.6, 4, 25.5, 2, 4.30, 1])
    bar_chart(data)


#// END_TODO [bar_chart]

data = simplified_splitser(5, [12.45, 1, 15.6, 4, 25.5, 2, 4.30, 1])
bar_chart(data)
````
